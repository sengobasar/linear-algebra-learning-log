## Example
example = $R^2 \to R^3$ (Feature Expansion)

**Conceptual Compression**

This transformation lifts data into a higher-dimensional space.
No information is destroyed; new features are created as linear combinations.

Used when:

- adding features
- embedding lower-dimensional data
- intermediate layers in neural networks

```python
import numpy as np

A = np.array([[0, 1],
              [1, 1],
              [1, -1]])

v = np.array([2, 3])

print(A @ v)
```
