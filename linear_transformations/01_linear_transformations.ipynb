# Conceptual Compression — Linear Transformations

A linear transformation is a transformation $T: R^n \to R^m$ satisfying

$$T(u+v) = T(u) + T(v)$$

$$T(cu) = cT(u)$$

for all vectors $u, v$ in $R^n$ and all scalars $c$.

**What this really means**

- Add first or transform first → same result
- Scale first or transform first → same result

So the transformation respects everything you learned about vectors.

```python
# Example
import numpy as np

# A linear transformation (matrix)
A = np.array([[2, 0],
              [0, 1]])

u = np.array([1, 2])
v = np.array([3, 1])
a = 3

print("T(u + v):", A @ (u + v))
print("T(u) + T(v):", A @ u + A @ v)

print("T(a·u):", A @ (a * u))
print("a·T(u):", a * (A @ u))
```

A linear transformation is a function between vector spaces:

$$L: V \to W$$

It must satisfy two conditions:

**(1) Scaling is preserved**
$$L(cv) = cL(v)$$

**(2) Addition is preserved**
$$L(v+w) = L(v) + L(w)$$
